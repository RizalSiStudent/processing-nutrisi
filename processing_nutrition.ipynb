```python
# Baca dataset
df = pd.read_csv("nutrition.csv")

# Melihat nama kolom agar tahu struktur dataset
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Filter data untuk makanan Abon dan Ayam
abon_df = df[df['name'].str.contains('Abon', case=False, na=False)]
ayam_df = df[df['name'].str.contains('Ayam', case=False, na=False)]
```

```python
# Ambil rata-rata nilai nutrisi dari masing-masing kelompok
abon_mean = abon_df[['calories', 'fat', 'carbohydrate', 'proteins']].mean()
ayam_mean = ayam_df[['calories', 'fat', 'carbohydrate', 'proteins']].mean()
```

```python
# Gabungkan dalam satu DataFrame untuk perbandingan
compare_df = pd.DataFrame({'Abon': abon_mean, 'Ayam': ayam_mean})
```

```python
# Plot perbandingan
compare_df.plot(kind='bar', figsize=(8,5))
plt.title('Perbandingan Kandungan Gizi: Abon vs Ayam')
plt.ylabel('Jumlah (gram atau kkal)')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Makanan')
plt.show()
```

```python
# Ambil 5 makanan dengan kalori tertinggi
top5_calories = df.nlargest(5, 'calories')[['name', 'calories', 'fat', 'carbohydrate', 'proteins']]

# Plot grafik perbandingan kandungan gizi 5 makanan tersebut
top5_calories.set_index('name')[['calories', 'fat', 'carbohydrate', 'proteins']].plot(kind='bar', figsize=(10,6))
plt.title('Top 5 Makanan dengan Kalori Tertinggi dan Kandungan Gizinya')
plt.ylabel('Jumlah (gram atau kkal)')
plt.xticks(rotation=30, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Kandungan Gizi')
plt.tight_layout()
plt.show()
```

```python
foods = ['Abon','Ayam','Telur','Ikan','Tahu']
gizi = df[df['name'].isin(foods)].groupby('name')[['calories','fat','carbohydrate','proteins']].sum()

plt.pie(gizi['calories'], labels=gizi.index, autopct='%1.1f%%',
        startangle=140, wedgeprops={'width':0.4})
plt.title('Perbandingan Kalori 5 Makanan Populer')
plt.show()
```

```python
foods = ['Abon','Ayam','Tahu']
gizi = df[df['name'].isin(foods)][['name','calories','fat','carbohydrate','proteins']].set_index('name')

gizi.plot(kind='line', marker='o', figsize=(10,6))
plt.title('Perbandingan Kandungan Gizi 3 Makanan Populer (Abon, Ayam, Tahu)')
plt.ylabel('Jumlah (gram atau kkal)')
plt.xticks(rotation=0)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
